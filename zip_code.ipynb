```python
def pars(i):
    url='https://kaz.postcodebase.com/ru/node/'+str(i)#4177
    a=UserAgent().chrome
    header={'user-agent':a}
    t=requests.get(url,headers=header)
    #t.status_code
    soup=BeautifulSoup(t.text,'html.parser')
    gorod=soup.find_all("span", attrs={ "itemprop" : "addressLocality"})[0].text
    region1=soup.find_all("span", attrs={ "itemprop" : "addressRegion"})[0].text
    region2=soup.find_all("span", attrs={ "itemprop" : "addressRegion"})[1].text
    country=soup.find_all("span", attrs={ "itemprop" : "addressCountry"})[0].text
    zip_code=soup.find_all("span", attrs={ "itemprop" : "postalCode"})[0].text
    
    return gorod,region1,region2,country,zip_code,url
    
```

```python
%%time
gorod_list_ru=[]
gorod_list_en=[]
region1_list_ru=[]
region1_list_en=[]
region2_list_ru=[]
region2_list_en=[]
country_list=[]
zip_code_list=[]
url_list=[]

for i in tqdm(range(1,10)):
    gorod_tmp,region1_tmp,region2_tmp,country_tmp,zip_code_tmp,url_tmp=pars(i)
    gorod_list_ru.append(gorod_tmp.split('/')[0])
    gorod_list_en.append(gorod_tmp.split('/')[1])
    region1_list_ru.append(region1_tmp.split('/')[0])
    region1_list_en.append(region1_tmp.split('/')[1])
    region2_list_ru.append(region2_tmp.split('/')[0])
    region2_list_en.append(region2_tmp.split('/')[1])
    country_list.append(country_tmp)
    zip_code_list.append(zip_code_tmp)
    url_list.append(url_tmp)
    
```

```python
df=pd.DataFrame({ 'gorod_ru':gorod_list_ru,
                  'gorod_en':gorod_list_en,
                  'region1_ru':region1_list_ru,
                  'region1_en':region1_list_en,
                  'region2_ru':region2_list_ru,
                  'region2_en':region2_list_en,
                  'country':country_list,
                  'zip_code':zip_code_list,
                  'url':url_list,
            })
```
