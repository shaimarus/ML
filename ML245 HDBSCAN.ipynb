### Подключение библиотек

```python
import pandas as pd
import numpy as np
import hdbscan
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
```

### Загрузка данных

```python
train_data = pd.read_csv('https://video.ittensive.com/machine-learning/hacktherealty/exposition_train.basic.csv.gz')
train_data.head()
```

### Нормализация данных
Приведение всех значений к отрезку [0;1]

```python
train_data_mm = pd.DataFrame(MinMaxScaler().fit_transform(train_data))
```

```python
train_data_mm.head()
```

### Иерархический DBSCAN
https://hdbscan.readthedocs.io/en/latest/api.html

```python
hdb = hdbscan.HDBSCAN(min_cluster_size=5)
hdb.fit(train_data_mm)
```

```python
print ("Неразмеченные данные:", str(round(100*len(hdb.labels_[(hdb.labels_==-1)&(hdb.probabilities_>0.5)])/len(train_data_mm)))+"%")
```

```python
print(1 - metrics.davies_bouldin_score(train_data_mm, hdb.labels_)) # -0.5
```

Элементы первого кластера

```python
train_data.iloc[list(np.where((hdb.labels_==0)&(hdb.probabilities_>0.5))[0])]
```
