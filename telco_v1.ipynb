```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\1\\1\\telco-customer-churn.csv')
```

```python
plt.figure(figsize=(12,6))
ax=sns.countplot(x='gender',hue='churn',data=df)
ax.set_title('gender vs churn',fontsize=20)
ax.set_xlabel('',fontsize=15)
ax.set_ylabel('number of gender',fontsize=15)

plt.figure(figsize=(12,6))
ax=sns.countplot(x='education',hue='churn',data=df)
ax.set_title('gender vs churn',fontsize=20)
ax.set_xlabel('',fontsize=15)
ax.set_ylabel('number of gender',fontsize=15)
```

```python
plt.figure(figsize=(12,6))
ax=sns.barplot(x='education',y='churn',data=df)
ax.set_title('gender vs churn',fontsize=20)
ax.set_xlabel('',fontsize=15)
ax.set_ylabel('number of gender',fontsize=15)
```

```python
plt.figure(figsize=(12,6))
ax=sns.boxplot(x='churn',y='gender',data=df)
ax.set_title('gender vs churn',fontsize=20)
ax.set_xlabel('churn',fontsize=15)
ax.set_ylabel('number of gender',fontsize=15)
```

```python
agg=df[['numberofmonthunpaid','numberofcomplaints','churn']].groupby('churn',as_index=False)[['numberofmonthunpaid','numberofcomplaints']].sum()
ax=agg.set_index('churn').T.plot(kind='bar',stacked=True,figsize=(12,6))
patches,labels=ax.get_legend_handles_labels()
ax.legend(patches,labels,loc='best')
ax.set_title('sdfsdfsdf',fontsize=20)
```

```python
import matplotlib.ticker as mtick
ax=(df['churn'].value_counts()*100.0/len(df)).plot(kind='bar',stacked=True,rot=0,figsize=(8,6))
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_ylabel('%, Customers', size=14)
ax.set_xlabel('Churn',size=14)
ax.set_title('Churn Rate',size=14)

totals=[]

for i in ax.patches:
    totals.append(i.get_width())

total=sum(totals)

for i in ax.patches:
    ax.text(i.get_x()+.15, i.get_height()-4.0,\
            str(round((i.get_height()/total),1))+'%',fontsize=12,color='white',weight='bold',size=14)
```

```python
agg=df[['numberofmonthunpaid','churn']].groupby('churn',as_index=False)[['numberofmonthunpaid']].sum()
ax=agg.set_index('churn').T.plot(kind='bar',stacked=True,figsize=(12,6))
patches,labels=ax.get_legend_handles_labels()
ax.legend(patches,labels,loc='best')
ax.set_title('sdfsdfsdf',fontsize=20)
```

```python
agg=df.groupby(['education','churn']).size().unstack()
ax=(agg.T*100.0/agg.T.sum()).T.plot(kind='bar',width=0.3,stacked=True,rot=0,figsize=(12,6))
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_ylabel('%, Customers', size=14)
ax.set_xlabel('numberofmonthunpaid',size=14)
ax.set_title('numberofmonthunpaid Rate',size=14)

for p in ax.patches:
    width,height=p.get_width(),p.get_height()
    x,y=p.get_xy()
    ax.annotate('{:.0f}%'.format(height),(p.get_x()+.25*width,p.get_y()+.4*height),color='white',weight='bold',size=14)
```
